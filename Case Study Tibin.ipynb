```python
#3.Plot a bar graph of male vs female buyers participated in the sales.
df['Buyer Gender'].value_counts()
```

```python
df['Buyer Gender'].value_counts().plot(kind='bar');


plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title("male vs female participated in sales", y=1.02);
plt.grid()
plt.show(block=True);
```

```python
#4.Find the least 5 cars based on their Resell price.

least_car=df[['Make','Model', 'Nickname','Resell Price']]
resell_list=least_car.sort_values('Resell Price')
resell_list[:5]
```

```python
#5. Find the least 5 cars based on their Sale price.
sale_pri=df[['Make','Model','Nickname', 'Sale Price']]

sale_price=sale_pri.sort_values('Sale Price',ascending=False)
sale_price[:5]
```
